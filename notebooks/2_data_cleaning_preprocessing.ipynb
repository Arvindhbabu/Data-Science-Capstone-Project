## Step 2: Preprocessing

```python
import pandas as pd

df = pd.read_csv("raw_spacex_launches.csv")

print("Shape of raw data:", df.shape)
df.head()
```

## Step 3: Data Cleaning

```python
# Remove empty and irrelevant columns
df = df.dropna(axis=1, how='all')  # Drop fully empty columns
df = df.dropna(axis=0, how='all')  # Drop fully empty rows

# Reset column names if multi-indexed or weirdly named
df.columns = df.columns.map(str)

# Display basic info
df.info()

# Try displaying unique values in 'Landing outcome' or similar columns
for col in df.columns:
    print(f"\nColumn: {col}\n", df[col].unique())
```

## Step 4: Feature Engineering

```python
def map_landing(outcome):
    if isinstance(outcome, str) and "Success" in outcome:
        return 1
    elif isinstance(outcome, str) and "Failure" in outcome:
        return 0
    else:
        return None

df['Landing_Success'] = df['Booster landing'].apply(map_landing)
```

## Step 5: Save Cleaned Data

```python
df.to_csv("cleaned_spacex_launches.csv", index=False)

files.download("cleaned_spacex_launches.csv")
```
